```python
# Deal with dataframes
import pandas as pd
import numpy as np

# Creating fake data
from faker import Faker
fake = Faker()
Faker.seed(12345)
```

```python
def generate_fake_summary_table():
    fake = Faker()
    list_account_basic_info = []
    for i in range(1,12001):
        account_id = fake.uuid4()
        total_users = fake.random_int(min=1, max = 20)
        mrr_ticket = fake.random_int(min = 50, max = 1000)
        list_account_basic_info.append((account_id, total_users, mrr_ticket))
        dataframe_name = pd.DataFrame(list_account_basic_info, columns=('AccountId',
                                                                              'TotalUsers',
                                                                              'MRR'))

    return dataframe_name     
```

```python
def generate_fake_users_complaints_per_account():
    list_complaints = []
    for i in df_account_basic_info.index:
        account_id = df_account_basic_info.iloc[i]['AccountId']
        num_complaint_users = fake.random_int(min = 0, max = df_account_basic_info.iloc[i]['TotalUsers'])
        if num_complaint_users == 0:
            num_complaints = 0
        else:
            num_complaints = fake.random_int(min= num_complaint_users, max = 20)
        alert_description = 'Complaints'
        list_complaints.append((account_id, num_complaint_users, num_complaints, alert_description))
        dataframe_name = pd.DataFrame(list_complaints, columns=('AccountId',
                                                                'TotalUsersWithComplaints',
                                                                'TotalComplaints',
                                                                'AlertDescription'))
    return dataframe_name
```

```python
def generate_fake_users_usage():
    list_complaints = []
    for i in df_account_basic_info.index:
        account_id = df_account_basic_info.iloc[i]['AccountId']
        num_usage_users = fake.random_int(min = 0, max = df_account_basic_info.iloc[i]['TotalUsers'])
        if num_usage_users == 0:
            monthly_total_saas_usage = 0
        else:
            monthly_total_saas_usage = fake.random_int(min= num_usage_users, max = 100)
        list_complaints.append((account_id, num_usage_users, monthly_total_saas_usage))
        dataframe_name = pd.DataFrame(list_complaints, columns=('AccountId',
                                                                'MAU',
                                                                'MonthlyTotalSaasUsage'))
    return dataframe_name
```

```python
def generate_fake_users_test_product():
    list_complaints = []
    for i in df_account_basic_info.index:
        account_id = df_account_basic_info.iloc[i]['AccountId']
        num_users_tested_new_feature = fake.random_int(min = 0 , max = df_monthly_users.iloc[i]['MAU'])
        list_complaints.append((account_id, num_users_tested_new_feature))
        dataframe_name = pd.DataFrame(list_complaints, columns=('AccountId',
                                                                'NumUsersTestedFeature'))
    return dataframe_name
```

### Aggregated Bases

```python
df_complaints = df_account_basic_info[['AccountId', 'TotalUsers']].merge(df_users_complaints, how='left', on='AccountId')
df_complaints['UsersWithComplaintsTotalUsersPerc'] = (df_complaints['TotalUsersWithComplaints']/df_complaints['TotalUsers'])*100
df_complaints = df_complaints.fillna(0)
df_complaints = df_complaints.rename(columns = {'TotalUsers': 'QttUserReference',
                                                'TotalUsersWithComplaints':'QttUserWithAlerts',
                                                'TotalComplaints':'TotalAlerts',
                                                'UsersWithComplaintsTotalUsersPerc':'UsersWithAlertPerc'})
df_complaints
```

```python

df_overall_usage = df_account_basic_info[['AccountId', 'TotalUsers']].merge(df_monthly_users, how='left', on='AccountId')
df_overall_usage['ProdUsageTotalUsersPerc'] = (df_overall_usage['MAU']/df_overall_usage['TotalUsers'])*100
df_overall_usage['AlertDescription'] = 'Overall Product Usage'
df_overall_usage = df_overall_usage.fillna(0)
df_overall_usage = df_overall_usage.rename(columns = {'TotalUsers': 'QttUserReference',
                                                'MAU':'QttUserWithAlerts',
                                                'MonthlyTotalSaasUsage':'TotalAlerts',
                                                'ProdUsageTotalUsersPerc':'UsersWithAlertPerc'})
df_overall_usage
```

```python
df_new_feature_usage = df_monthly_users[['AccountId', 'MAU']].merge(df_new_feature_tested, how='left', on='AccountId')
df_new_feature_usage['NewFeatureProductUsagePerc'] = (df_new_feature_usage['NumUsersTestedFeature']/df_new_feature_usage['MAU'])*100
df_new_feature_usage = df_new_feature_usage.fillna(0)
df_new_feature_usage['NotUsed'] = np.where(df_new_feature_usage['NumUsersTestedFeature'] > 0, 0, 1)
df_new_feature_usage['AlertDescription'] = 'New Feature Usage'
df_new_feature_usage = df_new_feature_usage.rename(columns = {'MAU': 'QttUserReference',
                                                'NumUsersTestedFeature':'QttUserWithAlerts',
                                                'NotUsed':'TotalAlerts',
                                                'NewFeatureProductUsagePerc':'UsersWithAlertPerc'})
df_new_feature_usage
```

```python
df_alerts = pd.concat([df_complaints, df_overall_usage, df_new_feature_usage])
df_alerts
```

```python
df_alerts_full = df_account_basic_info.merge(df_alerts, how='left', on='AccountId')
df_alerts_full
```
